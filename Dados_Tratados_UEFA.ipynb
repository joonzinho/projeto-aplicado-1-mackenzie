```
#Importando bibliotecas

import pandas as pd
```

```
#Criando dataframe

df_players = pd.read_excel("UEFA Champions League 2016-2022 Data.xlsx", sheet_name="players")
display(df_players)
```

```
#Criando dataframe

df_matches = pd.read_excel("UEFA Champions League 2016-2022 Data.xlsx", sheet_name="matches")
display(df_matches)
```

```
#Criando dataframe

df_goals = pd.read_excel("UEFA Champions League 2016-2022 Data.xlsx", sheet_name="goals")
display(df_goals)
```

```
#Renomeando a coluna "PID"

df_goals.rename(columns = {'PID':'PLAYER_ID'}, inplace=True)
display(df_goals)
```

```
#Excluindo colunas do dataframe df_players

df_players.drop(['NATIONALITY', 'DOB', 'JERSEY_NUMBER'], axis = 1, inplace=True)
display(df_players)
```

```
#Excluindo colunas do dataframe df_matches

df_matches.drop(['DATE_TIME', 'PENALTY_SHOOT_OUT'], axis = 1, inplace=True)
display(df_matches)
```

```
#Juntado dataframes com merge

df_goals_matches = pd.merge(df_goals, df_matches, how = 'inner', on = 'MATCH_ID')
"""display(df_goals_matches)"""
df_full = pd.merge(df_goals_matches, df_players, how = 'inner', on = 'PLAYER_ID')
display(df_full)
```

```
#Excluindo colunas do dataframe df_full

df_full.drop(['GOAL_ID', 'MATCH_ID', 'PLAYER_ID'], axis = 1, inplace=True)
display(df_full)
```

```
#Concatenando colunas "FIRST_NAME" e "LAST_NAME"

df_full['MARCADOR'] = df_full['FIRST_NAME'].map(str) + ' ' + df_full['LAST_NAME'].map(str)
display(df_full)
```

```
#Excluindo colunas "FIRST_NAME" e "LAST_NAME"

df_full.drop(['FIRST_NAME', 'LAST_NAME'], axis = 1, inplace=True)
display(df_full)
```

```
df_full = df_full[['SEASON', 'HOME_TEAM', 'AWAY_TEAM', 'HOME_TEAM_SCORE', 'AWAY_TEAM_SCORE', 'MARCADOR', 'DURATION', 'POSITION', 'GOAL_DESC','STADIUM', 'ATTENDANCE']]
display(df_full)
```
